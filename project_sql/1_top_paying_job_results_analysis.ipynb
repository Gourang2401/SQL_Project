```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data1 = pd.read_csv(r"C:\Users\Sristi v\Downloads\SQL_Project\project_sql\1_top_paying_jobs_results.csv")
```

```python
print(data1.head())
```

```python
df = data1.drop(['job_id','job_posted_date'],axis = 1 ) # dropping few unnecessary columns
```

```python
df1 = df.sort_values('salary_year_avg', inplace = False)
```

### Bar chart (Job titles vs Average salary)


```python
plt.figure(figsize=(15, 6))
plt.barh(df1['job_title'],df1['salary_year_avg'])
plt.title('Average salary distribution of top 10 paying Data analysts in India')
plt.xlabel('Average salary')
plt.ylabel('Job title')
plt.tight_layout()
plt.show()
```
